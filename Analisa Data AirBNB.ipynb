```python
#menampilkan rata-rata harga antara jenis kamar
df.groupby('room_type')['price'].mean()
```

```python
#menampilkan jumlah presentase penginapan disetiap location
location = ('Manhattan', 'Brooklyn', 'Queens', 'Bronx', 'Staten Islan')
votes = (21661, 20104, 5666, 1091, 373)
```

```python
colors = ('#8B4513', '#FFF8DC', '#93C572', '#E67F0D','#D53032')
```

```python
explode = (0, 0, 0.3, 0, 0)
```

```python
plt.pie(votes, 
        labels=location,
       autopct='%1.2f%%',
       colors=colors,
       explode = explode,
       shadow = True)
plt.title('Neighbour Group in AirBnb')
plt.show()
```

```python
# Menghitung rata-rata harga berdasarkan tipe kamar
room_type_prices = df.groupby('room_type')['price'].mean()

# Plotting bar chart

room_type_prices.plot(kind='bar')
plt.xlabel('Room Type')
plt.ylabel('Rata-rata Harga')
plt.title('Hubungan Room Type dengan Harga')
plt.show()
```

```python
# menampilkan hubungan harga dengan jumlah ulasan
plt.plot(df['price'], df['number_of_reviews'], 'bo')
plt.xlabel('Price')
plt.ylabel('Jumlah Ulasan')
plt.title('Relationship between Price and Reviews')
plt.grid(True)
plt.show()
```

```python
#Jumlah Penginapan berdasarkan tipe kamar
plt.figure(figsize=(8, 5))
sns.countplot(x='room_type', data=df)
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.title('Distribution of Room Types')
plt.xticks(rotation=45)
plt.show()
```

```python
#hubungan harga dan lokasi
plt.figure(figsize=(8, 5))
sns.boxplot(x='neighbourhood_group', y='price', data=df)
plt.xlabel('Neighbourhood Group')
plt.ylabel('Price')
plt.title('Price Distribution by Neighbourhood Group')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Menampilkan hubungan jumlah penginapan berdasarkan lokasi
count_by_neighbourhood = df['neighbourhood_group'].value_counts()

# Plotting bar chart
plt.figure(figsize=(8, 5))
count_by_neighbourhood.plot(kind='bar')
plt.xlabel('Lokasi')
plt.ylabel('Jumlah Penginapan')
plt.title('Jumlah Penginapan berdasarkan Lokasi')
plt.xticks(rotation=45)
plt.show()
```

```python
# menampilkan hubungan rata-rata harga berdasarkan lokasi
mean_price_by_neighbourhood = df.groupby('neighbourhood_group')['price'].mean()

# Plotting bar chart
plt.figure(figsize=(8, 5))
mean_price_by_neighbourhood.plot(kind='bar')
plt.xlabel('Lokasi')
plt.ylabel('Rata-rata Harga')
plt.title('Rata-rata Harga Penginapan berdasarkan Lokasi')
plt.xticks(rotation=45)
plt.show()
```

```python
# menampilkan ketersediaan dan harga penginapan
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='availability_365', y='price', hue='neighbourhood_group')
plt.xlabel('Availability (in days)')
plt.ylabel('Price')
plt.title('Relationship between Availability and Price by Neighbourhood Group')
plt.grid(True)
plt.show()
```

```python
#menampilkan hubungan jumlah review berdasarkan lokasi penginapan
plt.scatter(df['neighbourhood_group'], df['number_of_reviews'])

plt.title('Hubungan Lokasi Penginapan dengan Jumlah Review')
plt.xlabel('Lokasi Penginapan')
plt.ylabel('Jumlah Review')
plt.show()
```
